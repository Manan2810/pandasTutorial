```python
# skip rows and header have the same work
# df=pd.read_csv("reportcard.csv",skiprows=1)
df=pd.read_csv("reportcard.csv",header=1)
df
```

```python
# missing header
# if you dont have the column names given
df=pd.read_csv("reportcard.csv",header=None)#this lines give the column name as 0,1,2,3,..,etc
df
# for giving them names of your choice you can do as follows
df=pd.read_csv("reportcard.csv",header=None,names=["dates","events","temperature","windspeed"])
df
```

```python
# if given a large csv file and want to read only first three rows -> here we use nrows
df=pd.read_csv("reportcard.csv",nrows=3)
df
```

```python
# replacing n/a values with NAN - here we use na_values
df=pd.read_csv("reportcard.csv",na_values=["not available","n.a"])
df
```

```python
df=pd.read_csv("reportcard.csv",na_values=["not available","n.a"])
df
# here we want to make specifically windspeed NAN where its value = -1 so we will use dictionary for this purpose
df=pd.read_csv("reportcard.csv",na_values={
    "event":["not available","n.a"],
    "temperature":["not available","n.a"],
    "windspeed":["not available","n.a",-1]
})
df
```
